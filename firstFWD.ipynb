```python
def filter_day(df):
    days = ['Friday' , 'Saturday' , 'Sunday' , 'Monday' , 'Tuesday' , 'Wednesday' , 'Thursday']
    while True:
        day = input("what is the day? (Friday , Saturday , Sunday , Monday , Tuesday , Wednesday , Thursday) \n").title()
        if day in days: 
            df = df[df['day'] == day]            
            break
        else:
            print('note valed day ' )  
    return df
```

```python
def filter_month(df):
    months = ['January','February','March','April','May','June','July','August','September','October','November','December']
    month_in = df['month'].unique()   
    while True:
        month = input("what is the month? (January, February, March, April, May, June, July, August, September, October, November, December) \n").title()
        if month in month_in: 
            df = df[df['month'] == month]
            break
        elif month in  months:
            print('no trip in this month avelepol month are : ' ,month_in)
        else:
            print('note valed month ')  
    return df
```

```python
#import libraries
import pandas as pd
def city():  
    #let user choose the city
    city_files = { 'chicago': 'chicago.csv','new york': 'new_york_city.csv','washington': 'washington.csv' }
    while True: 
        x = input("Enter your city (Chicago, New York, Washington):\n").lower()
        if x in city_files.keys(): 
            df=pd.read_csv(city_files[x]) #make csv dataframe
            df['Start Time'] = pd.to_datetime(df['Start Time']) #transfer start time column from string to date time object
            df['month'] = df['Start Time'].dt.month_name() #make month column
            df['day'] = df['Start Time'].dt.day_name()  #make day column
            break
        else:
            print('Not a valid city')
    return df , x   #return Data Frame of city data and name of city


def filter_day(df):
    #filter data by day ,take df as dataframe 
    days = ['Friday' , 'Saturday' , 'Sunday' , 'Monday' , 'Tuesday' , 'Wednesday' , 'Thursday']
    while True:
        day = input("Enter the day (Friday , Saturday , Sunday , Monday , Tuesday , Wednesday , Thursday):\n").title()
        if day in days: 
            df = df[df['day'] == day]            
            break
        else:
            print('Not a valid day' )  
    return df #return df filterd by day

def filter_month(df):
    #filter data by day ,take df as dataframe 
    months = ['January','February','March','April','May','June','July','August','September','October','November','December']
    month_in = df['month'].unique()   
    while True:
        month = input("Enter the month (January, February, March, April, May, June, July, August, September, October, November, December):\n").title()
        if month in month_in: 
            df = df[df['month'] == month]
            break
        elif month in  months:
            print('no trip in this month avelepol month are : ' ,month_in)
        else:
            print('Not a valid month')  
    return df  #return df filterd by day

def filter_both(df):
    #filter data by day and month ,take df as dataframe 
    months = ['January','February','March','April','May','June','July','August','September','October','November','December']
    month_in = df['month'].unique()   
    while True:
        month = input("Enter the month (January, February, March, April, May, June, July, August, September, October, November, December):\n").title()
        if month in month_in: 
            days = ['Friday' , 'Saturday' , 'Sunday' , 'Monday' , 'Tuesday' , 'Wednesday' , 'Thursday']
            while True:
                day = input("Enter the day (Friday , Saturday , Sunday , Monday , Tuesday , Wednesday , Thursday):\n").title()
                if day in days: 
                    df = df[df['day'] == day]            
                    break
                else:
                    print('Not a valid day' )  
            df = df[df['month'] == month]
            break
        elif month in  months:
            print('No trips in this month, the available months are:  ' ,month_in)
        else:
            print('Not a valid month')  
    return df  #return df filterd by day and month

def most(df,x):
    #Calculate statistics
    print('Popular times of travel')
    print('Most commen month:' , df['month'].mode()[0])
    print('Most commen day:' , df['day'].mode()[0])
    print('Most commen hour:' , df['Start Time'].dt.hour.mode()[0] ,'\n')
    print('Popular stations and trip')
    print('Most commen Start Station:' , df['Start Station'].mode()[0])
    print('Most commen End Station:' , df['End Station'].mode()[0])
    print('Most commen trip:' , df.groupby(['Start Station','End Station']).size().idxmax() ,'\n')
    print('Trip duratione')
    print('Total travel time:',df['Trip Duration'].sum())    
    print('Average travel time:',df['Trip Duration'].mean() ,'\n')
    print('User info')
    print('Counts of each user type:\n',df['User Type'].value_counts().to_string())
    if x != 'washington': # washington data don't have gender or year of birth data so we don't Calculate it
        print('counts of each gender:\n',df['Gender'].value_counts().to_string())
        print('Earliest year of birth:', int(df['Birth Year'].min()))
        print('Recent year of birth:', int(df['Birth Year'].max()))
        print('Most common year of birth:', int(df['Birth Year'].mode()[0]))
    else:
        print('No gender data or year of birth in Washington')
        

        
        
def show(df):
    y=5
    while True:
        x=input('Do you want to show 5 rows of data ?(yes , no)\n').lower()
        if x == 'yes' and y < len(df):
            print(df[(y-5):y])
            y+=5
        elif x == 'no' or y > len(df):
            if x == 'yes':
                print(df[(y-5):len(df)])
                print('The data is end')
            print('Thanks')
            break
        else:
            print('Not valid choice')
        
        
def main():
    print('welcome!')
    df , x = city()
    while True:
        y = input("How you need to filter data (day, month, both, none):\n").lower()
        if y == 'day':
            df = filter_day(df) 
            most(df,x)
            show(df)
            break
        elif y == 'month':
            df = filter_month(df)
            most(df,x)
            show(df)
            break
        elif y == 'both':
            df = filter_both(df)
            most(df,x)
            show(df)
            break
        elif y == 'none':
            df = df
            most(df,x)
            show(df)
            break            
        else:
            print('Not valid choice')    
            
            
if __name__ == "__main__":
    main()
               
```
